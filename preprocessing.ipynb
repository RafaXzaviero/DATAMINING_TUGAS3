```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
```

```python
# Langkah 1: Memuat Data
# Ganti dengan path ke dataset Anda jika menggunakan dataset lain
data = {'Country': ['France', 'Spain', 'Germany', 'Spain', 'Germany', 'France', 'Spain', 'France', 'Germany', 'France'],
        'Age': [44, 48, 30, 45, 40, 35, 38, 48, 50, 37],
        'Salary': [72000, 48000, 54000, 61000, 58000, 52000, 79000, 83000, 67000, 69000],
        'Purchased': ['No', 'Yes', 'No', 'No', 'Yes', 'Yes', 'No', 'Yes', 'No', 'Yes']}
df = pd.DataFrame(data)

display(df)
```

```python
# Langkah 2: Menangani Nilai yang Hilang (jika ada)
# Menggunakan SimpleImputer untuk mengisi nilai NaN dengan mean (rata-rata) dari kolom
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')

# Terapkan imputer pada kolom 'Age' dan 'Salary'
# Pastikan kolom yang akan diimputasi adalah numerik
df[['Age', 'Salary']] = imputer.fit_transform(df[['Age', 'Salary']])

display(df)
```

```python
# Langkah 3: Mengkodekan Data Kategorikal
# One-Hot Encoding untuk kolom 'Country'
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
X = np.array(ct.fit_transform(df))

# Label Encoding untuk kolom 'Purchased' (variabel dependen)
le = LabelEncoder()
y = le.fit_transform(df['Purchased'])

print("Data setelah One-Hot Encoding (X):")
display(X)
print("\nData setelah Label Encoding (y):")
display(y)
```

```python
# Langkah 4: Penskalaan Fitur (Feature Scaling)
# Terapkan StandardScaler pada kolom numerik (kolom Age dan Salary setelah One-Hot Encoding)
# Perhatikan indeks kolom setelah One-Hot Encoding: kolom Age adalah kolom ke-3 dan Salary adalah kolom ke-4
# Jika menggunakan data yang berbeda, pastikan untuk menyesuaikan indeks kolom
scaler = StandardScaler()
X[:, 3:5] = scaler.fit_transform(X[:, 3:5])

print("Data setelah Feature Scaling (X):")
display(X)
```

```python
# Langkah 5: Membagi Data menjadi Training Set dan Test Set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

print("Bentuk X_train:", X_train.shape)
print("Bentuk X_test:", X_test.shape)
print("Bentuk y_train:", y_train.shape)
print("Bentuk y_test:", y_test.shape)
```

Variabel X: Ini adalah variabel yang berisi fitur atau variabel independen dari dataset Anda. Setelah melalui tahapan preprocessing:
Kolom kategorikal 'Country' telah diubah menjadi format numerik menggunakan One-Hot Encoding. Ini berarti satu kolom kategorikal 'Country' dipecah menjadi beberapa kolom biner (0 atau 1), masing-masing merepresentasikan salah satu kategori negara (France, Spain, Germany). Ini dilakukan agar algoritma machine learning dapat memprosesnya.
Kolom numerik 'Age' dan 'Salary' telah mengalami Penskalaan Fitur (Feature Scaling) menggunakan StandardScaler. Ini mengubah nilai-nilai pada kolom tersebut sehingga memiliki rata-rata 0 dan standar deviasi 1. Penskalaan ini membantu mencegah fitur dengan nilai yang lebih besar mendominasi algoritma machine learning.
Kolom asli 'Purchased' (sebelum di-encode menjadi y) juga masih ada di sini karena kita menggunakan remainder='passthrough' di ColumnTransformer, namun biasanya kolom target ini akan dipisahkan sebelum penskalaan fitur. Dalam konteks machine learning, X akan digunakan sebagai input untuk melatih model.
Variabel y: Ini adalah variabel yang berisi target atau variabel dependen dari dataset Anda. Kolom 'Purchased' yang awalnya berisi nilai kategorikal ('Yes' dan 'No') telah diubah menjadi nilai numerik biner (0 dan 1) menggunakan Label Encoding. Ini adalah format yang dibutuhkan oleh sebagian besar algoritma machine learning untuk variabel target dalam tugas klasifikasi. Dalam konteks machine learning, y adalah apa yang ingin diprediksi oleh model berdasarkan fitur-fitur di X.
Singkatnya, X adalah data input yang sudah diproses dan diskalakan untuk model, sementara y adalah output target yang sudah diubah menjadi format numerik.
